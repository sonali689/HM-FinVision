### KOTAKBANK AND TATASTEEL

#### KOTAKBANK is stationary and TATASTEEL is non-stationary

```python
import yfinance as yf
import pandas as pd
from statsmodels.tsa.stattools import adfuller, kpss
import matplotlib.pyplot as plt

# ADF Test
def adf_test(series):
    result = adfuller(series, autolag='AIC')
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    for key, value in result[4].items():
        print(f'Critical Values: {key}, {value}')
    return result[1] 

# KPSS Test
def kpss_test(series):
    result = kpss(series, regression='ct', nlags="auto")
    print(f'KPSS Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    for key, value in result[3].items():
        print(f'Critical Values: {key}, {value}')
    return result[1] 

def get_stock_data(ticker, start, end):
    stock_data = yf.download(ticker, start=start, end=end)
    stock_data = stock_data['Close'] 
    stock_data.dropna(inplace=True)
    return stock_data

if __name__ == "__main__":

    stock_1 = "KOTAKBANK.NS" 
    stock_2 = "TATASTEEL.NS" 
    start_date = "2023-01-01"
    end_date = "2024-12-01"


    print("\nFetching data for Stock 1...")
    stock_1_data = get_stock_data(stock_1, start_date, end_date)
    print("\nFetching data for Stock 2...")
    stock_2_data = get_stock_data(stock_2, start_date, end_date)


    plt.figure(figsize=(14, 7))
    plt.plot(stock_1_data, label=stock_1)
    plt.plot(stock_2_data, label=stock_2)
    plt.title("Stock Prices")
    plt.xlabel("Date")
    plt.ylabel("Close Price")
    plt.legend()
    plt.show()

    print(f"\nPerforming ADF and KPSS tests for {stock_1}...")
    print(f"{stock_1} - ADF Test:")
    adf_test(stock_1_data)
    print(f"{stock_1} - KPSS Test:")
    kpss_test(stock_1_data)

    print(f"\nPerforming ADF and KPSS tests for {stock_2}...")
    print(f"{stock_2} - ADF Test:")
    adf_test(stock_2_data)
    print(f"{stock_2} - KPSS Test:")
    kpss_test(stock_2_data)
```
